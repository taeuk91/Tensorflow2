```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
# x, y축 만들기
df = pd.DataFrame(columns = ['x', 'y'])
```

```python

# x, y에 랜덤한 수를 30번 넣기
for i in range(101):
  df.loc[i] = [random.randint(0, 100), random.randint(0, 100)]
```

```python
df.head(30)
```

```python
# 데이터 시각화 하기
# x, y좌표 형태로, data frame을 점의 크기 100으로 출력)
sb.lmplot('x', 'y', data=df, fit_reg=False, scatter_kws={"s": 100})
# label 넣기
plt.title('K-means Eample')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# 특정 명시 없이 자동으로 K-means++ 알고리즘 수행 -> 총 4개의 클러스터 중심 위치 찾기
points = df.values
kmeans = KMeans(n_clusters=4).fit(points)
kmeans.cluster_centers_
```

```python
# 데이터 확인
kmeans.labels_
```

```python
# 30개의 데이터가 어떤 클러스터에 속하는지 확인
df['cluster'] = kmeans.labels_
df.head(30)
```

```python
# 데이터 시각화 하기
# x, y좌표 형태로, data frame을 점의 크기 100으로 출력)
sb.lmplot('x', 'y', data=df, fit_reg=False, scatter_kws={"s": 150}, hue = "cluster")
# label 넣기
plt.title('K-means Eample')
plt.xlabel('x')
plt.ylabel('y')
```
